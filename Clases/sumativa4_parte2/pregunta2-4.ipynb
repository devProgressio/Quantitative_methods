```python
from sympy import symbols, Eq, solve

# Definir la variable simbólica k
k = symbols('k')

# Definir la ecuación de la suma de probabilidades igual a 1
ecuacion = Eq(0.25*k + 0.8*k + 3*k + 2*k + 0.4 + 0.1 + 0.03, 1)

# Resolver la ecuación para k
solucion_k = solve(ecuacion, k)

# Mostrar el valor de k
print(solucion_k)
```

```python
import matplotlib.pyplot as plt

# Definir los valores de x y las probabilidades correspondientes
valores_x = [0, 1, 2, 3, 4, 5, 6]
probabilidades = [0.25 * solucion_k[0], 0.4, 0.8 * solucion_k[0], 0.1, 3 * solucion_k[0], 0.03, 2 * solucion_k[0]]

# Crear el gráfico de barras
plt.bar(valores_x, probabilidades, align='center', alpha=0.7)

# Etiquetas y título del gráfico
plt.xlabel('Número de seguros de vida vendidos')
plt.ylabel('Probabilidad')
plt.title('Distribución de la variable aleatoria')

# Mostrar la gráfica
plt.show()
```

```python
from math import comb

# Número de ejecutivos
n_ejecutivos = 15

# Número de ejecutivos que no venden seguros
k_no_venden = 5

# Probabilidad de que un ejecutivo no venda seguros (P(X=0))
p_no_venden = 0.25 * solucion_k[0]

# Calcular la probabilidad utilizando la distribución binomial
probabilidad = comb(n_ejecutivos, k_no_venden) * (p_no_venden ** k_no_venden) * ((1 - p_no_venden) ** (n_ejecutivos - k_no_venden))
# Mostrar el resultado
print(f"La probabilidad de que 5 de los 15 ejecutivos no vendan seguros es: {probabilidad:.20f}")
```

```python
# Definir el valor esperado (media) del modelo, que es el primer momento crudo
valor_esperado = sum(x * prob for x, prob in zip(valores_x, probabilidades))

# Calcular la probabilidad de que X > 6
probabilidad_mas_de_6 = 1 - sum(prob for x, prob in zip(valores_x[:7], probabilidades[:7]))

# Mostrar el resultado
print(f"La probabilidad de que un ejecutivo venda más de 6 seguros en un día es: {probabilidad_mas_de_6:.20f}")
```
