```python
import numpy as np

# Definir los valores de X y sus respectivas probabilidades
X = np.array([0, 1, 2, 3, 4, 5])
P_X = np.array([0.05, 0.1, 0.4, 0.3, 0.1, 0.05])

# 1. Calcular la probabilidad de que se realice a lo más 3 veces
prob_at_most_3 = np.sum(P_X[X <= 3])

# 2. Calcular el valor esperado (promedio)
promedio = np.sum(X * P_X)

# 3. Calcular la desviación estándar
desviacion_estandar = np.sqrt(np.sum((X - promedio) ** 2 * P_X))

# Imprimir resultados con dos decimales
print(f"1. La probabilidad de que se realice a lo más 3 veces es: {prob_at_most_3:.2f}")
print(f"2. El promedio es: {promedio:.2f}")
print(f"3. La desviación estándar es: {desviacion_estandar:.2f}")
```

```python
import matplotlib.pyplot as plt
# Definir los valores de X y sus respectivas probabilidades
X = np.array([0, 1, 2, 3, 4, 5])
P_X = np.array([0.05, 0.1, 0.4, 0.3, 0.1, 0.05])

# Crear el gráfico de barras
plt.bar(X, P_X, color='blue', alpha=0.7)

# Añadir etiquetas y título
plt.xlabel('X (Número de repeticiones diarias)')
plt.ylabel('P(X)')
plt.title('Gráfico de la función de probabilidad - A.Pérez')

# Guardar el gráfico como un archivo PNG
plt.savefig('grafico_probabilidad_aperez.png')

# Mostrar el gráfico
plt.show()
```

```python
from scipy.integrate import quad

# Definir la función de densidad de probabilidad (pdf)
def pdf(x):
    if x > 0:
        return 20000 / (x + 100)**3
    else:
        return 0

# Calcular la probabilidad de que la vida útil sea al menos 200 días
prob_at_least_200, _ = quad(pdf, 200, float('inf'))

# Calcular la probabilidad de que la vida útil esté entre 80 y 120 días
prob_between_80_and_120, _ = quad(pdf, 80, 120)

# Calcular la vida útil esperada (valor esperado)
esperanza, _ = quad(lambda x: x * pdf(x), 0, float('inf'))

# Imprimir resultados con dos decimales
print(f"1. La probabilidad de que la vida útil sea al menos 200 días es: {prob_at_least_200:.2f}")
print(f"2. La probabilidad de que la vida útil esté entre 80 y 120 días es: {prob_between_80_and_120:.2f}")
print(f"3. La vida útil esperada es: {esperanza:.2f} días")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de densidad de probabilidad (pdf)
def pdf(x):
    return np.where(x > 0, 20000 / (x + 100)**3, 0)

# Generar valores de x para el gráfico
x_values = np.linspace(0, 200, 1000)

# Calcular los valores correspondientes de la función de densidad de probabilidad
y_values = pdf(x_values)

# Crear el gráfico
plt.plot(x_values, y_values, label='Función de Densidad de Probabilidad')
plt.title('Función de Densidad de Probabilidad - aperez')
plt.xlabel('Vida Útil (días)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

# Guardar el gráfico como un archivo PNG
plt.savefig('grafico_densidad_probabilidad_aperez.png')

# Mostrar el gráfico
plt.show()
```
